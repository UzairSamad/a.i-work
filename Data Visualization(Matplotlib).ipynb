# Plotting Graphs with matplotlib

## Simple line graph with x and y axis labels

```python
import matplotlib.pyplot as plt
x = range(1,51) ## generating range foe x
y = [i**2 for i in x ]   ## generating values for y
plt.plot(x,y, label="First Line")

x1 = [10,200,4]
y1 = [40,50,400]
plt.plot(x1,y1, label="Second Line")


plt.title("Simple Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-Axis")
plt.legend()
plt.show()
```

## another line graph

```python
## line 1
x1 = [10,20,30]
y1 = [20,40,10]

## line 2
x2 = [10,20,30]
y2 = [40,10,30]

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")



plt.plot(x1,y1, label= "First Line with width 3" , color='red', linewidth="3")
plt.plot(x2,y2, label="Second Line with width 6" , color = 'blue' , linewidth='6')

plt.title("Two lines with different colors and width")
plt.legend()
plt.show()
```

## Bar charts

```python
import matplotlib.pyplot as plt

x = [12,20,30,40,50,60,70]
y = [100,130,160,190,210,240,270]

y1= [150,20,30,40,50,60,70]
x1 = [14,10,10,41,24,67,77]

plt.bar(x,y, label="bar1", color='red' , linewidth="6")
plt.bar(x1,y1, label="bar2", color='black', linewidth='5')
plt.xlabel("X-Axis")
plt.title("A simple bar chart")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x=["JavaScript","Python" ,"C++","Java","Php"]
y = [72.5,80,25,75,50 ]

plt.bar(x,y)
plt.xlabel("name of languages")
plt.ylabel("percentage of like")
plt.title("Ratios of languages developer like")

plt.show()
```

```python
import matplotlib.pyplot as plt

x=["JavaScript","Python" ,"C++","Java","Php"]
y = [72.5,80,25,75,50 ]

plt.barh(x,y)                          ## for horizontal we use plt.barh
plt.xlabel("name of languages")
plt.ylabel("percentage of like")
plt.title("Ratios of languages developer like")

plt.show()
```

```python
import matplotlib.pyplot as plt
x=["JavaScript","Python" ,"C++","Java","Php"]
y = [72.5,80,25,75,50 ]
x_new = [i for i, _ in enumerate(x)]
plt.barh(x_new,y, color='green' )

plt.xlabel("popularity")
plt.ylabel("ratio")
plt.title("languages developer likes the most")
plt.yticks(x_new,x)

##turn on the grid
plt.minorticks_on()
plt.grid(which='major',linestyle='-',linewidth='0.6', color='red')

## minor grid
plt.grid(which='minor' ,linestyle='-', linewidth='0.3', color='black')
plt.show()


```

```python
import matplotlib.pyplot as plt

population_ages=[22,33,44,33,22,55,66,77,88,87,98,33,22]

bins=[0,10,25,35,50,60,70,80,100,150]


plt.hist(population_ages,bins, histtype='bar', rwidth=0.8, label='Ages')
plt.xlabel("X label")
plt.ylabel("Y label")

plt.title("Histogram")
plt.legend()
plt.show()
```

# Scater Plot

```python
import matplotlib.pyplot as plt

x = [1,1.5,3,5,7,9,4,9,8,11,3]
y= [2,8,4,6,8,7,9,10,4,12,2]

plt.scatter(x,y , label ='Example',color='red' , marker='*')
plt.legend()
plt.title("Scatter Plot")
plt.ylabel('Y')
plt.xlabel('X')
plt.show()
```

# StackPlot

```python
days=[1,2,3,4,5]

reading=[3,6,8,9,9]
playing=[4,5,9,3,4]
watching=[13,7,6,8,12]
teaching=[2,15,20,4,5]


plt.plot([],[], color='c', label='days', linewidth=5)
plt.plot([],[], color='m', label='reading', linewidth=5)
plt.plot([],[], color='r', label='watching', linewidth=5)
plt.plot([],[], color='k', label='playing', linewidth=5)
plt.plot([],[], color='g', label='teaching', linewidth=5)
plt.stackplot(days,reading,playing,watching,teaching , colors=['c','m','r','k','g'])

plt.legend()
plt.title('Scatter Plot')
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.show()
```
